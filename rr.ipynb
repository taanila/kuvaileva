```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Piilotan mahdolliset varoitukset
import warnings
warnings.filterwarnings("ignore")

plt.style.use('seaborn-whitegrid')

# Kaikki rivit ja sarakkeet näytetään
pd.options.display.max_rows = None
pd.options.display.max_columns = None
```

```python
df = pd.read_excel('https://taanila.fi/rahastoraportti.xlsx')
df.info()
```

```python
df.head()
```

```python
# 12 kuukauden tuottoprosentin suhteen top 10 rahastot
df.nlargest(columns = '12M_return', n = 10)
```

```python
# Eri rahastoyhtiöiden hallinnoimien rahastojen lukumäärät
df1 = pd.crosstab(df['company'], 'n').sort_values(by = 'n', ascending = False)
df1.columns.name = ''
df1
```

```python
# Eri rahastoluokkiin kuuluvien rahastojen lukumäärät
df2 = pd.crosstab(df['RR2009_class'], 'n').sort_values(by = 'n', ascending = False)
df2.columns.name = ''
df2
```

```python
# Eri aikavälien tuottoprosenttien tunnusluvut
df[['120M_return', '60M_return', '36M_return',
       '12M_return', '6M_return', '3M_return', '1M_return']].describe()
```

```python
# 12 kuukauden, 6 kuukauden ja 1 kuukauden tuottoprosenttien jakaumat
sns.boxplot(data = df[['12M_return', '6M_return', '1M_return']] * 100, orient = 'h')
plt.xlabel('Tuottoprosentti')
```

```python
# Muodostan datan, jossa mukana vain suomalaisiin osakkeisiin sijoittavat rahastot
df3 = df[df['RR2009_class'] == 'EQ_FINLAND']
df3
```

```python
# Suomalaisiin osakkeisiin sijoittavien rahastojen 12 kk tuottojen tunnuslukuja
df3['12M_return'].describe()
```

```python
# Suomalaisiin osakkeisiin sijoittavien rahastojen 12 kuukauden tuottojen luokiteltu jakauma
bins = [0, 10, 20, 30, 40]
df3['12M_luokiteltu'] = pd.cut(df3['12M_return'] * 100, bins = bins)

df4 = pd.crosstab(df3['12M_luokiteltu'], 'Rahastojen lkm')
df4.columns.name = ''
df4.index.name = '12 kk tuotto %'
df4
```

```python
# Edellinen pylväinä
df4.plot.barh()
```
